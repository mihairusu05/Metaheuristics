```python
MAX_COORD = 5.12
MIN_COORD = -5.12

MIN_COORD_S = -65.536
MAX_COORD_S = 65.536
```

```python
def sphere_fitness_function(individual) :
    """Fitness function for the sphere problem

    Args:
        individual (list[float]): solution

    Returns:
        (float): fitness of solution
    """
    return sum(x**2 for x in individual)
```

```python
def schwefel_fitness_function(individual : list[float]) -> float :
    """Fitness function for the schwefel problem

    Args:
        individual (list[float]): solution

    Returns:
        sum (float): fitness of solution
    """
    sum = 0
    
    for i in range(len(individual)) :
        for j in range(i+1) :
            sum+=individual[j]**2
            
    return sum    
```

```python
import random

def pso1_for_sphere(swarm_size: int = 30, dimension_size: int = 3,
                   nr_of_iterations: int = 100,
                   w: float = 0.7, c1: float = 1.6, c2: float = 1.6):

    # Init population and p_bests
    swarm_of_particles = []
    fitness_of_swarm_of_particles = []

    p_swarm_of_particles = []
    fitness_p_of_swarm_of_particles = []

    best_fitness = float('inf')
    best_particle = None

    velocities = [] 

    for _ in range(swarm_size):
        particle = [random.uniform(MIN_COORD, MAX_COORD) for _ in range(dimension_size)]
        v = [random.uniform(MIN_COORD, MAX_COORD) for _ in range(dimension_size)]

        velocities.append(v)

        fitness = sphere_fitness_function(particle)  

        if fitness < best_fitness:
            best_particle = particle.copy()
            best_fitness = fitness

        swarm_of_particles.append(particle)
        fitness_of_swarm_of_particles.append(fitness)

        p_swarm_of_particles.append(particle.copy())         
        fitness_p_of_swarm_of_particles.append(fitness)      

    
    best_global_particle = p_swarm_of_particles[0].copy()
    best_global_fitness = fitness_p_of_swarm_of_particles[0]

    for i in range(len(fitness_p_of_swarm_of_particles)):
        if fitness_p_of_swarm_of_particles[i] < best_global_fitness:
            best_global_particle = p_swarm_of_particles[i].copy()
            best_global_fitness = fitness_p_of_swarm_of_particles[i]

    for _ in range(nr_of_iterations):
        for i in range(swarm_size):

            r1 = random.random()
            r2 = random.random()

        
            velocities[i] = [
                w * velocities[i][d]
                + c1 * r1 * (p_swarm_of_particles[i][d] - swarm_of_particles[i][d])
                + c2 * r2 * (best_global_particle[d] - swarm_of_particles[i][d])
                for d in range(dimension_size)
            ]

            swarm_of_particles[i] = [
                swarm_of_particles[i][d] + velocities[i][d]
                for d in range(dimension_size)
            ]

            swarm_of_particles[i] = [
                max(MIN_COORD, min(MAX_COORD, swarm_of_particles[i][d]))
                for d in range(dimension_size)
            ]

            fitness_of_swarm_of_particles[i] = sphere_fitness_function(swarm_of_particles[i])

            if fitness_of_swarm_of_particles[i] < fitness_p_of_swarm_of_particles[i]:
                p_swarm_of_particles[i] = swarm_of_particles[i].copy()
                fitness_p_of_swarm_of_particles[i] = fitness_of_swarm_of_particles[i]

            if fitness_p_of_swarm_of_particles[i] < best_global_fitness:
                best_global_particle = p_swarm_of_particles[i].copy()
                best_global_fitness = fitness_p_of_swarm_of_particles[i]

    return best_global_particle, best_global_fitness
```

```python
def pso2_for_sphere(swarm_size: int = 30, dimension_size: int = 3,
                                 nr_of_iterations: int = 100,
                                 c1: float = 2.05, c2: float = 2.05):

    phi = c1 + c2
    chi = 2 / abs(2 - phi - (phi**2 - 4*phi)**0.5)

    velocities = []
    swarm_of_particles = []
    fitness_of_swarm = []
    p_best_particles = []
    fitness_p_best = []

    best_global_particle = None
    best_global_fitness = float('inf')

    for _ in range(swarm_size):
        particle = [random.uniform(MIN_COORD, MAX_COORD) for _ in range(dimension_size)]
        v = [random.uniform(-abs(MAX_COORD - MIN_COORD), abs(MAX_COORD - MIN_COORD)) for _ in range(dimension_size)]

        fitness = sphere_fitness_function(particle)

        velocities.append(v)
        swarm_of_particles.append(particle)
        fitness_of_swarm.append(fitness)
        p_best_particles.append(particle.copy())
        fitness_p_best.append(fitness)

        if fitness < best_global_fitness:
            best_global_particle = particle.copy()
            best_global_fitness = fitness

    for _ in range(nr_of_iterations):
        for i in range(swarm_size):

            r1 = random.random()
            r2 = random.random()

            velocities[i] = [
                chi * (
                    velocities[i][d]
                    + c1 * r1 * (p_best_particles[i][d] - swarm_of_particles[i][d])
                    + c2 * r2 * (best_global_particle[d] - swarm_of_particles[i][d])
                )
                for d in range(dimension_size)
            ]

            swarm_of_particles[i] = [
                swarm_of_particles[i][d] + velocities[i][d]
                for d in range(dimension_size)
            ]

            swarm_of_particles[i] = [
                max(MIN_COORD, min(MAX_COORD, swarm_of_particles[i][d]))
                for d in range(dimension_size)
            ]

            fitness_of_swarm[i] = sphere_fitness_function(swarm_of_particles[i])

            if fitness_of_swarm[i] < fitness_p_best[i]:
                p_best_particles[i] = swarm_of_particles[i].copy()
                fitness_p_best[i] = fitness_of_swarm[i]

            if fitness_p_best[i] < best_global_fitness:
                best_global_particle = p_best_particles[i].copy()
                best_global_fitness = fitness_p_best[i]

    return best_global_particle, best_global_fitness
```
