```python
import math
import random

def load_tsp(filename: str) -> list:
    coords = []
    with open(filename) as f:
        for line in f:
            if line.strip() == 'NODE_COORD_SECTION':
                break
        for line in f:
            if line.strip() == 'EOF':
                break
            parts = line.split()
            x, y = float(parts[1]), float(parts[2])
            coords.append((x, y))
    return coords
```

```python
def euclidean_distance(a: tuple, b: tuple) -> float:
    return math.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

def tour_length(tour, dist ):
    n = len(tour)
    return sum(dist[tour[i]][tour[(i + 1) % n]] for i in range(n))
```

```python
def greedy_tsp(filename: str) -> tuple:
    """A solution for the tsp problem using a greedy method.
    The method consists of starting at a random node and always selecting the closest unvisited node
    

    Args:
        filename (str): _description_

    Returns:
        tuple: _description_
    """
    
    coords = load_tsp(filename)
    n = len(coords)
    start = random.randint(0,n-1)

    tour = [start]
    available = set(range(n))
    available.remove(start)
    current = start
    
    dists = [[euclidean_distance(coords[i], coords[j]) 
         for j in range(n)] 
         for i in range(n)]

    while available:
        nearest = None
        nearest_dist = float('inf')

        for node in available:
            dist = euclidean_distance(coords[current], coords[node])
            if dist < nearest_dist:
                nearest_dist = dist
                nearest = node        
        tour.append(nearest)
        available.remove(nearest)
        current = nearest

    length = tour_length(tour, dists)
    return tour, length, dists
```

```python
import random
import math

def simulated_annealing_tsp(filename: str, T_max: float = 10000, T_min: float = 0.1,alpha: float = 0.999,iterations_per_temp: int = 100):

    coords = load_tsp(filename)
    n = len(coords)

    _, _, dists = greedy_tsp(filename)

    c = list(range(n))
    random.shuffle(c)
    c_length = tour_length(c, dists)

    best = c.copy()
    best_length = c_length

    T = T_max
    t = 0

    while T > T_min:

        for _ in range(iterations_per_temp):

            i = random.randint(0, n - 2)
            j = random.randint(i + 1, n - 1)
            x = c.copy()
            x[i:j+1] = x[i:j+1][::-1]
            x_length = tour_length(x, dists)
            
            delta = x_length - c_length

            if delta < 0:
                c = x
                c_length = x_length
            elif random.random() < math.exp(-delta / T):
                c = x
                c_length = x_length

            if c_length < best_length:
                best = c.copy()
                best_length = c_length

        T = alpha * T
        t+=1
        if t% 250 == 0 :
            print(f"Result at step_{t} is {best_length}")

    return best_length
```
